**5)**. Y a-t-il des valeurs anormales dans l'échantillon ?

**6)**. Exprimez le nombre 13,7 en termes de moyenne et d'écart-type de ces données.

**Reponses :**

**4)**

La sample mean (moyenne) est : $\bar X = \frac{10 + 11 + 15 + 15 + 17 + 22}{5} = 15.0$

Le standard deviation (ecart-type) est : $s = \sqrt{\frac{(10-15)^2 + (11-15)^2 + (15-15)^2 + (15-15)^2 + (17-15)^2 + (22-15)^2)}{5}} = 4.34$

**5)** 

Par convention, les valeurs supérieures à 3 fois l'écart type par rapport à la moyenne sont considérées comme anormales. Calculons l'intervalle en dehors duquel les valeurs sont considérées comme anormales : $[\bar X - 3s,\bar X + 3s] = [15 - 3*4.34, 15 + 3*4.34] = [1.99, 28.01]$.

Il n'y a pas de données d'échantillon en dehors de cet intervalle, ce qui signifie qu'il n'y a pas de valeurs anormales dans l'échantillon.

**6)**

Pour répondre à cette question, nous pouvons calculer le **Z-score** : $Z = \frac{X - \bar X}{s} = \frac{13.7 - 15.0}{4.34} = -0.3$.

Cela signifie que la valeur $13.7$ est situé à $−0.3$ **écarts-types à gauche** de la moyenne.

La moyenne est :  15.0</br>
L'ecart-type est :  4.34</br>
Trois écarts-types en dessous de la moyenne:  1.99</br>
Trois écarts-types au dessus de la moyenne:  28.01</br>
Le **Z-score** pour 13.7 est :  -0.3

```python
sample = pd.Series([10, 11, 15, 15, 17, 22])
m = sample.mean()
std = sample.std()

X = 13.7
z = (X - m) / std

print(f"La moyenne est : {sample.mean()}")
print(f"L'ecart-type est : {round(sample.std(),2)}")
print(f"Trois écarts-types en dessous de la moyenne : {round(m - 3*std, 2)}")
print(f"Trois écarts-types au dessus de la moyenne : {round(m + 3*std, 2)}")
print(f"Le Z-score pour 13.7 est : {round(z,2)}")
```

### Exercise 2

Un supermarché a mené plusieurs campagnes publicitaires. Ils aimeraient analyser la relation entre le coût de la publicité et le chiffre d'affaires.

Laisser $X$ et $Y$ être le coût de la publicité et le chiffre d'affaires. Nous avons un échantillon de $n=10$ valeurs pour $X$ et $Y$ , trié de telle sorte que $X_i$ , $Y_i$ sont les résultats de la $𝑖$ -ème campagne marketing :

$X$  : 39, 65, 62, 90, 82, 75, 25, 98, 36, 78

$Y$  : 47, 53, 58, 86, 62, 68, 60, 91, 51, 84

**1)**. Calculer les moyennes de l'échantillon $\bar X$ et $\bar Y$

### $\bar X = \frac{39 + 65 + 62 + 90 + 82 + 75 + 25 + 98 + 36 + 78}{10} = 65$

### $\bar Y = \frac{47 + 53 + 58 + 86 + 62 + 68 + 60 + 91 + 51 + 84}{10} = 66$

```python
data = {'X' : [39, 65, 62, 90, 82, 75, 25, 98, 36, 78],
        'Y' : [47, 53, 58, 86, 62, 68, 60, 91, 51, 84]}

sample = pd.DataFrame(data)
print(sample.mean())
sample
```

**2)**. Calculer les écarts-types de l'échantillon $s_X$ et $s_Y$

Pour trouver l'écart-type, calculons la racine carrée des variances

### $s_X = \sqrt{\frac{(39-65)^2 + (65-65)^2 + (62-65)^2 + (90-65)^2 + (82-65)^2 + (75-65)^2 + (25-65)^2 + (98-65)^2 + (36-65)^2 + (78-65)^2}{10}}$

### $s_X = 24.49$

### $s_Y = \sqrt{\frac{(47-66)^2 + (53-66)^2 + (58-66)^2 + (86-66)^2 + (62-66)^2 + (68-66)^2 + (60-66)^2 + (91-66)^2 + (51-66)^2 + (84-66)^2}{10}}$

### $s_Y = 15.72$

```python
print("Sample standard deviations for X and Y :")
print(round(sample.std(), 2))
```

**3)**. Calculer la covariance entre $X$ et $Y$

La covariance est:
    
### $\textrm{Cov(X,Y)} = {\frac {1}{n}}\sum _{i=1}^{n}(X_{i} - \bar{X})(Y_{i} - \bar{Y})$

### $\textrm{Cov(X,Y)} = {\frac {1}{10}}[(39 - 65)(47 - 66) + (65 - 65)(53 - 66) + ... + (78 - 65)(84 - 66)]$

### $\textrm{Cov(X,Y)} = 300.44$

```python
# avec numpy

print(round(np.cov(sample['X'], sample['Y'])[0][1], 2))

# avec pandas

print(round(sample.cov().iloc[0,1], 2))
```

**4)**. Calculer le coefficient de corrélation de Pearson entre $X$ et $Y$

Le coefficient de corrélation de Pearson peut être déduit de la covariance et des écarts types

### $ \rho _{X,Y} = {\frac {\textrm{Cov} (X,Y)}{s _{X}s _{Y}}}$

### $ \rho _{X,Y} = 0.78$

```python
import scipy
from scipy import stats

r, p = scipy.stats.pearsonr(sample['X'], sample['Y'])

print("coefficient de corrélation de Pearson :", round(r, 2))
print("p-value :", round(p, 4))
```

```python
# avec numpy

r = np.corrcoef(sample['X'], sample['Y'])[0][1]

print(round(r, 2))
```

```python
# avec pandas

r = sample.corr().iloc[0,1]

print(round(r, 2))
```

Le coefficient de corrélation entre $X$ et $Y$ est positif et assez proche de 1, ce qui signifie qu'il existe une relation linéaire entre ces deux variables : lorsque X augmente, Y augmente également en général, proportionnellement à X

## Normal distribution and Central Limit Theorem

### Exercise 3

**1)**. Les tailles des femmes aux États-Unis sont normalement distribuées avec une moyenne de 64 pouces et un écart type de 2,75 pouces. Quel est le pourcentage de femmes dont la taille est comprise entre 64 et 69,5 pouces ?

L'intervalle $[64,69.5]$ Correspond à $[\mu, \mu + 2\sigma]$ où $\mu = 64$ est la hauteur moyenne et $\sigma = 2.75$ l'écart type.

Les hauteurs sont normalement distribuées, nous savons donc que $95 \%$ des femmes ont des hauteurs dans l'intervalle $[\mu - 2\sigma, \mu + 2\sigma]$.

Comme la distribution normale est symétrique, on peut en déduire que $\frac{95}{2} = 47.5 \%$ des femmes mesurent entre 64 et 69,5 pouces

**2)**. Vous tirez à plusieurs reprises des échantillons de taille 100 à partir d'une population avec une distribution uniforme continue sur l'intervalle $[0, 160]$ . La moyenne de cette population est $\mu=80$ . Pour chaque échantillon, vous calculez la somme $X$ des 100 valeurs.

a) Quelle est la répartition attendue de ces sommes $X$ ?

b) Quelle est la probabilité que la somme d'un échantillon soit inférieure à 8 000 ?

**a)**

Considérons la somme de chaque échantillon, $X$ . Selon le théorème central limite, les quantités qui sont des sommes de nombreuses mesures indépendantes sont presque normalement distribuées.

Ici, $X$ est la somme de 100 valeurs indépendantes, celle-ci est suffisamment élevée pour appliquer le théorème central limite. Ainsi, la distribution attendue de $X$ sera proche d'une distribution normale.

La moyenne de cette distribution normale sera $\bar X = 100 \times \mu = 100 \times 80 = 8000$

**b)**

Comme la distribution Normale est symétrique, on sait que 50% des valeurs sont situées en dessous de la moyenne $\bar X = 8000$ . En d'autres termes, la probabilité que la somme soit inférieure à 8000 est $P(X \le 8000) = 50 \%$

```python
import random

X = pd.Series(index = range(0, 1000), dtype='float')

# Repetons 1000 fois l'expérience

for i in range(0, 1000):
    # Créer un échantillon de 100 valeurs aléatoires uniformément réparties sur [0, 160]
    sample = pd.Series([random.uniform(0, 160) for i in range(100)])
    
    # Calculez la somme X des 100 valeurs de l'échantillon et enregistrez-la dans X
    X[i] = sample.sum()
    
# Plot la distribution de X
X.hist(bins=20)
```

$ E = z \frac{\sigma}{\sqrt{n}} = 1.645 \times \frac{4}{\sqrt{30}} = 1.20085$

L'intervalle de confiance à 90% est :

$ CI_{90\%} = [\mu - E, \mu + E] = [41 - 1.20085, 41 + 1.20085] = [39.80, 42.20]$
